```ruby
require_relative '../modules/aoc'

puzzle = Puzzle.new(2020,2) do | input |
    re = /([0-9]+)-([0-9]+) ([a-z]): (.+)/
    input.lines.map do | x | 
        m = x.match(re)
    end
end

puzzle.display(1)  
puzzle.display(2)
```

```ruby
def puzzle.valid1?(m)
    minocc = m[1].to_i
    maxocc = m[2].to_i
    c = m[3]
    ccount = m[4].count(c)
    (minocc..maxocc).include? ccount
end

def puzzle.valid2?(m)
    pos1 = m[1].to_i - 1
    pos2 = m[2].to_i - 1
    c = m[3]
    (m[4][pos1] == c) ^ (m[4][pos2] == c) 
end

def puzzle.part1
    @input.filter { |m| valid1?(m) }.count
end

def puzzle.part2
    @input.filter { |m| valid2?(m) }.count
end
puzzle.solve
```
