```elixir
defmodule AOC do
def digits (line) do
    digits = Regex.scan(~r/[0-9]/, line)
    [List.first(digits),List.last(digits)] 
    |> Enum.join("")
    |> String.to_integer
end
end

AOC.digits("")
```

```elixir
{:ok, input} = File.read("input/1.txt")
input
|> String.trim
|> String.split("\n")
|> Enum.map(&AOC.digits(&1))
|> Enum.sum
```
