```python
from pathlib import Path
import re
```

```python
%%capture
input = Path("input/1.txt").read_text().splitlines()
```

```python
re_digit = re.compile("[0-9]")

def first_digit(line):
    return re_digit.search(line).group(0)

def last_digit(line):
    return re_digit.search(line[::-1]).group(0)
    
```

```python
first_digit(input[0]) +  last_digit(input[0])
```

```python
sum([int(first_digit(line) + last_digit(line)) for line in input])
```
