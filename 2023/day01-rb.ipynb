# Part 1

```ruby
File.readlines("input/1.txt", chomp: true).map { | l |
    (l[/[0-9]/,0] + l.reverse[/[0-9]/,0]).to_i
}.sum
```

# Part 2

```ruby
re_part2 = "one|two|three|four|five|six|seven|eight|nine"
File.readlines("input/1.txt", chomp: true).map { | l |
    (l[/([0-9]|#{re_part2})/,1] + l.reverse[/([0-9]|#{re_part2.reverse})/,1].reverse)
}.map { | r | 
    r
    .gsub("one", "1")
    .gsub("two", "2")
    .gsub("three", "3")
    .gsub("four", "4")
    .gsub("five", "5")
    .gsub("six", "6")
    .gsub("seven", "7")
    .gsub("eight", "8")
    .gsub("nine","9")
}.map { | d | d.to_i }.sum
```
