```clojure
(defn fuel 
    [mass]
    (- (quot mass 3) 2))

(fuel 1969)
(def input (map #(Integer/parseInt %) (clojure.string/split-lines (slurp "../2019/input/1.txt"))))
```

```clojure
(apply + (map fuel input))
```

```clojure
(defn total-fuel 
    [mass] 
        (apply + (drop 1 (take-while (comp not neg?) (iterate fuel mass)))))
```

```clojure
(apply + (map total-fuel input))
```
