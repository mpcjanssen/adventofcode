```elixir
defmodule AOC do
    def fuel(mass) do
        div(mass,3) - 2
    end


    def total_fuel(mass) do
        Stream.iterate(mass, &fuel(&1)) 
        |> Stream.drop(1)
        |> Stream.take_while(&(&1 > 0))
        |> Enum.sum
    end
end
```

```elixir
{ok, contents} = File.read("./input/1.txt")
fuels = contents 
|> String.split("\n", trim: true)
|> Enum.map(&String.to_integer &1)

fuels |> Enum.map(&AOC.fuel &1)
|> Enum.sum
```

```elixir
fuels 
|> Enum.map(&AOC.total_fuel &1)
|> Enum.sum
```
