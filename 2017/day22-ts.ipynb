```typescript
import * as fs from 'fs';
```

```typescript
let input = fs.readFileSync("input/22.txt","utf-8").trim().split("\n").map(
    x => x.split("")
)
let gridsize = input.length
let start = (gridsize+1)/2
let pos = {x: start, y: start}
```

```typescript
let grid = new Set()
for (const [y, row] of input.entries()) {
    for (const [x, c] of row.entries()) {
        grid[`${x},${y}`] =  c;
    }
}
```

```typescript
let state = {
    grid: grid,
    pos: pos,
    dir: {dx: 0, dy: -1},
    bursts: 0
    
}
```

```typescript
function step(state) {
    let grid, pos, dir, bursts
    ({grid, pos, dir, bursts} = state)
    state.bursts = bursts + 1
    return state
}
step(state)
```
