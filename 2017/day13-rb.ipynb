```ruby
require_relative '../modules/aoc'

puzzle = Puzzle.new(2017,13) do | input |
    fws = {}
    input.lines.map do | l |
        l.chomp.split(": ").map(&:to_i)
    end.map do | depth, range | 
        fws[depth] = range 
    end
    fws
end

puzzle.display(1)  
puzzle.display(2)
```

```ruby
def puzzle.pos(range, ps)
    cycle = ((0...range).to_a + (1...range-1).to_a.reverse)
    posincycle = ps % (cycle.count)
    cycle[posincycle]
end

def puzzle.atzero(range, ps)
    cycle = 2*(range-1)
    return ps % cycle == 0
end

def puzzle.severity(delay)
maxdepth = @input.keys.max
severity = 0
caught = false
(0..maxdepth).each { 
    | depth |
    range = @input[depth]
#     p "no fw at depth #{depth}" unless range
    if range  && atzero(range,depth+delay)
#         p "Caught at #{depth}"
        severity += range * depth

    end
}
severity
end

def puzzle.caught(delay)
maxdepth = @input.keys.max
(0..maxdepth).each { 
    | depth |
    range = @input[depth]
#     p "no fw at depth #{depth}" unless range
    if range  && atzero(range,depth+delay)
#         p "Caught at #{depth}"
        return true
    end

}
return false
end

def puzzle.state(delay,ps)
    maxdepth = @input.keys.max
    (0..maxdepth).each { | x |
        range = @input[x]
        prefix = if x == ps 
            "*" 
        else 
            " " 
        end
        puts "#{prefix} #{x}: pos: #{pos(range,delay+ps)}" if range
        puts "#{prefix} #{x}: --" unless range
        
    }
end
```

```ruby
def puzzle.part1
    severity(0)
end

def puzzle.part2
    (0..10_000_000).each do | delay |
        break delay unless caught(delay)
    end
end
puzzle.solve
```
