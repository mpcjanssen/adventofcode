```ruby
require_relative '../modules/aoc'

puzzle = Puzzle.new(2017,16) do | input |
    dancers = ('a'..'p').to_a
    moves = input.chomp.split(',')
    start = dancers.clone
    count = 0
    dances = {}

    cyclecount = while true      
        moves.each { | move |
            case move
                when /^s([0-9]+)/
                    dancers.rotate!(-$1.to_i)
                when /^x([0-9]+)\/([0-9]+)/
                    a = $1.to_i
                    b = $2.to_i
                    dancers[a],dancers[b]=dancers[b],dancers[a]
                when /^p([a-z]+)\/([a-z]+)/
                    a = dancers.index($1)
                    b = dancers.index($2)
                    dancers[a],dancers[b]=dancers[b],dancers[a]
                else
                    raise "oops #{move}"
               end
        }
        count += 1
        dances[count] = dancers.clone
        break count if dancers == start
    end
    dances
end

puzzle.display(1)  
puzzle.display(2)
```

```ruby
def puzzle.part1
    @input[1].join
end

def puzzle.part2
    @input[1_000_000_000 % @input.count].join
end
puzzle.solve
```
