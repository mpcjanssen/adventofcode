````tcl
source ../modules/setup.tcl

set year 2017
set day 20

aoc::get-puzzle $year $day 1 
aoc::get-puzzle $year $day 2
set input [string trim [aoc::get-input $year $day]]
jupyter::display "text/markdown" "## Input
```
[string range $input 0 100]...
```";
````

````tcl
aoc::solve $input {
    set id -1
    set accs [lmap x [split $input \n] {
        regexp  {a=<(.*)>} $x -> accs ; list [incr id] [+ {*}[lmap y [split $accs ,] {expr {abs($y)}}]]}]
    yield [lindex [lsort -increasing -integer -index 1 $accs] 0 0]
}
````

````tcl
lmap part [split $input \n] {regexp {p=<([-0-9]+),([-0-9]+),([-0-9]+)>, } $part -> p1 p2 p3 ; list $p1 $p2 $p3 }
````
