````tcl
source ../modules/setup.tcl

set year 2017
set day 21

aoc::get-puzzle $year $day 1
# aoc::get-puzzle $year $day 1 2
set input [string trim [aoc::get-input $year $day]]
jupyter::display "text/markdown" "## Input
```
[string range $input 0 100]...
```";
````

````tcl
set patterns [lmap l [split  $input \n] {lindex [split $l " "] 0}]
proc pinfo {p} {
    set raw [string map [list "/" ""] $p]
    puts $raw
    string length $raw
    set lights [string length [string map {. ""} $raw]]
    puts $lights

}

lmap p $patterns {pinfo $p}
````
