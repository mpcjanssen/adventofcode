```python
def advent22b(s):
    rows = s.splitlines()
    map = {}
    h = len(rows)//2
    for x,r in enumerate(rows):
        for y,c in enumerate(r):
            if c == "#":
                map[y - x*1j] = "i" # up = decrease in x; right = increase in y
    dirc = 0 + 1j # up
    posc = h - h * 1j
    infections = 0
    for i in range(10000000):
        node = (map[posc] if posc in map else "c")
        if node == "c": 
            dirc *= 1j
            map[posc] = "w"
        elif node == "w":
            map[posc] = "i"
            infections += 1
        elif node == "i":
            dirc *= -1j
            map[posc] = "f"
        elif node == "f":
            dirc *= -1
            map[posc] = "c"
        posc += dirc
    return infections
```
