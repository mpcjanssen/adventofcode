```clojure
(import java.security.MessageDigest)
(defn md5
  [^String s]
  (->> s
       .getBytes
       (.digest (MessageDigest/getInstance "MD5"))
       (BigInteger. 1)
       (format "%032x")))
```

```clojure
(md5 "a")
```

```clojure
(time (take 1 
     (filter
      #(.startsWith % "00000")
      (map  
       #(md5 (str "ugkcyxxp" %)) 
        (range)))))
```

```clojure
(.startsWith "00000" "av")
```

```clojure
(.startsWith (md5 "abc3231929") "00000" )
```
