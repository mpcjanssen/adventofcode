```ruby
insts = input = File.readlines("./input/2.txt").map do | l |
    l.chomp!
    cmd, cnt = l.split
    [cmd, cnt.to_i]
end

state = {d: 0, f: 0} 
insts.each_with_object(state) do |inst, state|
    cmd, cnt = inst
    case cmd
        when "forward" 
            state[:f] += cnt
        when "down" 
            state[:d] += cnt
        when "up" 
            state[:d] -= cnt
    else throw cmd
    end
end
state[:d]*state[:f]
```

```ruby
state = {d: 0, f: 0, aim: 0} 
insts.each_with_object(state) do |inst, state|
    cmd, cnt = inst
    case cmd
        when "forward" 
            state[:f] += cnt
            state[:d] += cnt*state[:aim]
        when "down" 
            state[:aim] += cnt
        when "up" 
            state[:aim] -= cnt
    else throw cmd
    end
end
state[:d]*state[:f]
```
