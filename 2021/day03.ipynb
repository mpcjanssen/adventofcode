### Solve today

`aoc::solve input body`:
    The body is executed as a coroutine. Input is available as the `$input` variable. The results are returned using `[yield]`

````tcl
proc prefixoverlap {w1 w2} {
    set overlap 0
    foreach l1 [split $w1 {}] l2 [split $w2 {}] {
        if {$l1 eq $l2} {
            incr overlap
        } else {
            return $overlap
        }
    }
    return $overlap
}

prefixoverlap 1010 10100
````

````tcl
proc bin2dec bin {
    if {$bin == 0} {
        return 0
    } elseif {[string match -* $bin]} {
        set sign -
        set bin [string range $bin[set bin {}] 1 end]
    } else {
        set sign {}
    }
    return $sign[expr 0b$bin]
}

set test {00100
11110
10110
10111
10101
01111
00111
11100
10000
11001
00010
01010}

aoc::solve $input {
    # $input is available in the body
    # return the results using yield
    set input [split $input \n]
    set digits [string length [lindex $input 0]]
    foreach l $input {
        set idx 0
        foreach d [split $l ""] {
            incr occ($idx,$d)
            incr idx
        }
    }
    set gamma {}
    set epsilon {}
    for {set idx 0} {$idx < $digits} {incr idx} {
        if {$occ($idx,0) > $occ($idx,1)} {
            append gamma 0
            append epsilon 1
        } else {
            append gamma 1
            append epsilon 0
        }
    }
    
    set result1 [* [bin2dec $gamma] [bin2dec $epsilon]]

    yield $result1
    set result2 0
    puts $gamma
    puts [lsort -index 0 -integer -decreasing [lmap l $input { list [prefixoverlap $l $gamma] $l }] ]
    puts $epsilon
    puts [lsort -index 0 -integer -decreasing [lmap l $input { list [prefixoverlap $l $epsilon] $l }] ]
    yield $result2
} 
````
