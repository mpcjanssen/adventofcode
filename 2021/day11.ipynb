### Solve today

`aoc::solve input body`:
    The body is executed as a coroutine. Input is available as the `$input` variable. The results are returned using `[yield]`

````tcl
set test {5483143223
2745854711
5264556173
6141336146
6357385478
4167524645
2176841721
6882881134
4846848554
5283751526}
proc step grid {
    set grid [dict map {pos v} $grid {incr v}]
    while {true} {
        set flashed false
        foreach {flashspot _} [dict filter $grid script {k v}  { expr {$v ne "x" && $v > 9 }}]  {
            set flashed true
            # puts $flashspot
            dict set grid $flashspot x
            foreach n [aoc::neighbours8 {*}$flashspot] {
                if {[dict exists $grid $n] && [dict get $grid $n] ne "x"} {
                    dict incr grid $n
                }
            }
            # puts $grid
        }
        if {!$flashed} break
    }
    set flashes [dict filter $grid value x]
    set all false
    if {$flashes eq $grid} {
        set all true
    }
    set cflash 0
    foreach {p v} $flashes {dict set grid $p 0 ; incr cflash}
    return [list $cflash $grid $all]
}
````

````tcl
aoc::solve $input {
    set grid [aoc::todict $input]
    set total_flashes 0
    for {set step 0} true {} {
        lassign [step $grid] flashes grid all
        incr step
        incr total_flashes $flashes
        if {$step == 100} {
            yield $total_flashes
        }
        if {$all} {
            yield $step
            break
        }
    } 
} 
````
