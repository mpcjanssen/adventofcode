```kotlin
@file:DependsOn("com.github.petitparser:java-petitparser:2.3.1")

import java.io.File
import org.petitparser.parser.primitive.CharacterParser.*
```

```kotlin
val input = File("./input/2.txt").readText()
```

```kotlin
val space = of(' ')
val command = letter().plus().flatten().seq(space.seq(digit().plus().flatten().map {it: String -> it.toInt()}))
val newline = of('\n')

val script = command.seq(newline.optional()).plus()
```

```kotlin
val parsed = script.parse(input)
parsed
```
