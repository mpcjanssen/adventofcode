```tcl
(require racket)
```

```tcl
(define input (map string->number (file->lines "input/1.txt")))
```

```tcl
(define pairs (for/list ([i1 input] [i2 (list-tail input 1)]) (list i1 i2)))
(length (filter (lambda (p) (apply < p)) pairs))
```

```tcl
(define 3sums (for/list ([i1 input] [i2 (list-tail input 1)] [i3 (list-tail input 2)]) (+ i1 i2 i3)))
(define pairs (for/list ([i1 3sums] [i2 (list-tail 3sums 1)]) (list i1 i2)))
(length (filter (lambda (p) (apply < p)) pairs))
```
