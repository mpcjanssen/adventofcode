````tcl
array unset -nocomplain visited
set start {0 0}
set head $start
set tail $start
set visited($tail) 1
foreach move [split {R 4
U 4
L 3
D 1
R 4
D 1
L 5
R 2} \n] {
   lassign $move dir dist
   puts "Move $dir by $dist"
   lassign $head hx hy
   switch $dir {
        R {incr hx $dist}
        L {incr hx -$dist}
        U {incr hy -$dist}
        D {incr hy $dist}
   }
   set head [list $hx $hy]
   puts "Head at $head"
}
````
