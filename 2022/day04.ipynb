### Solve today

`aoc::solve input body`:
    The body is executed as a coroutine. Input is available as the `$input` variable. The results are returned using `[yield]`

````tcl
proc contains? {s1 s2} {
    set s1 [split $s1 -]
    set s2 [split $s2 -]
    lassign $s1 s1s s1e
    lassign $s2 s2s s2e
    expr {($s1s <= $s2s) && ($s1e >= $s2e)}
}

proc overlap? {s1 s2} {
    set s1 [split $s1 -]
    set s2 [split $s2 -]
    lassign $s1 s1s s1e
    lassign $s2 s2s s2e
    set s1tos2 [expr {(($s1e <= $s2e) && ($s1e >= $s2s)) || (($s1s <= $s2e) && ($s1s >= $s2s))}]
     set s2tos1 [expr {(($s2e <= $s1e) && ($s2e >= $s1s)) || (($s2s <= $s1e) && ($s2s >= $s1s))}]
     expr {$s1tos2 || $s2tos1}
}

````

````tcl
aoc::solve $input {
    # $input is available in the body
    # return the results using yield

    yield [+ {*}[lmap pair [split $input \n] {lassign [split $pair ,] p1 p2 ; expr {[contains? $p1 $p2] || [contains? $p2 $p1]}}]]
    yield [+ {*}[lmap pair [split $input \n] {lassign [split $pair ,] p1 p2 ; overlap? $p2 $p1}]]
} 
````
