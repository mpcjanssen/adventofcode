```elixir
{:ok, input} = File.read("./input/1.txt")
floors = input |> 
    String.codepoints |> 
    Enum.reduce([0], fn(ch,acc) -> 
        case {ch, acc} do
            {"(", [hd | _]} -> [hd+1 | acc]
            {")", [hd | _]} -> [hd-1 | acc]
        end
    end )
part1 = floors |> List.first
part2 = floors |> Enum.reverse |> Enum.find_index(fn(x) -> x == -1 end)
{part1, part2}
```
