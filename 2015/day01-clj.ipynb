```clojure
(def input (str (slurp "input/1.txt")))
(defn delta [x]
    (if (= x \() 1 -1))

(defn first-basement [idx floor deltas]
    (if (= -1 floor)
        idx
        (recur 
         (inc idx)
         (+ floor (first deltas)) 
         (rest deltas))))
(defn append-new-floor [v delta]
    (let [floor (last v)]
        (conj v (+ floor delta))))


(time (do
(def deltas (->>
    input
    (map delta)))

(def part1 (apply + deltas))
(def part2 (first-basement 0 0 deltas))
          ))
[part1 part2]
```
