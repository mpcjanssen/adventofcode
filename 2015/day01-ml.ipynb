```ocaml
let line = (let ic = open_in "input/1.txt" in
    input_line ic);;

let floors = 
    let nextfloor state ch =
        match (ch, state) with
           ('(', (idx, flr) :: _) ->  (idx + 1, flr +1) :: state
         | (')', (idx, flr) :: _) ->  (idx + 1, flr  - 1) :: state
         | (_, _) -> assert false in

    String.fold_left 
        nextfloor
        [(0, 0)] 
        line
    
    

let (_, part1) = floors |> List.hd

let (part2, _) = floors 
|> List.filter  (fun x -> let (idx, flr)  = x in flr = -1) 
|> List.rev
|> List.hd ;;

(part1, part2)
```
