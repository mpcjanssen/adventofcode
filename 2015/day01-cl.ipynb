```common-lisp
(ql:quickload :uiop)
(ql:quickload :cl-ppcre)
```

```common-lisp
(defparameter *input* (uiop:read-file-string "input/1.txt"))
```

```common-lisp
(let*  (
        (parts (cl-ppcre:split "\\(\\)" (sort *input* #'char-lessp)))
        (ups (first parts))
        (downs (second parts))) 
        (- (length ups) (length downs)))
```

```common-lisp
(defun floorchange (c) 
    (case c
        (#\( 1)
        (#\) -1)))
(map nil  #'floorchange  *input* )
```

```common-lisp
(floorchange #\))
```
