```python
from pathlib import Path
txt = Path('input/1.txt').read_text()
moves = [1 if c=='(' else -1 for c in txt]
```

```python
sum(moves)
```

```python
def floors(moves):
    floor = 0
    for m in moves:
        floor = floor + m
        yield floor
        
        
list(floors(moves)).index(-1)+1
```
