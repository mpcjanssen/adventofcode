```clojure
(def input (str (slurp "input/3.txt")))
```

```clojure
(defn move [pos dir]
     (let 
         [[x y] pos]
         (case dir
         \< [(dec x) y]
         \^ [x  (dec y)]
         \> [(inc x) y]
         \v [x (inc y)])))

(defn step [state d]
    (let [
          {p :p v :v } state
          n (move p d )
          ]
        {:p n :v (conj v n)}))    
```

```clojure
(def state {:p [0 0] :v #{} })
(def part1 (count (:v (reduce step state input))))
```
