```clojure
(require '[clojure.string :as str])
(def input (str (slurp "input/2.txt")))

(defn dimensions [box]
    (->> 
        (str/split box #"x") 
        (map #(Integer/parseInt %))))

(def boxes 
    (->>
        (str/split input #"\n")
        (map dimensions)))

(defn wrapping [d]
    (+
     ;; Slack
     (reduce * (take 2 (sort d)))
     ;; Box
     (* (reduce + (map #(reduce * %) (combo/combinations d 2) )) 2)))

(def part1
  (reduce + 
 (->> 
     boxes
     (map wrapping))))
```

```clojure
(defn bow [d]
    (+
     ;; Perimeter
     (* (reduce + (take 2 (sort d))) 2)
     ;; Box
     (reduce * d)))

(def part2
  (reduce + 
 (->> 
     boxes
     (map bow))))
```

```clojure
[part1 part2]
```
