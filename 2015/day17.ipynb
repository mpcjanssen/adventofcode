```tcl
##noresult
source ../modules/setup.tcl
set year 2015
set day 17
aoc::get-puzzle $year $day 1
aoc::get-puzzle $year $day 2
set input [string trim [aoc::get-input $year $day]]
jupyter::display text/plain [string range $input 0 200]...;
```

```tcl
aoc::solve $input {
    set l [llength $input]
    # puts $l
    set fits 0
    set fitsmin 0
    set result2 {}
    for {set i 1 } {$i <= $l} {incr i} {
        foreach comb [aoc::combinations $input $i] {
            if {[+ {*}$comb] == 150} {
                incr fits
                if {$result2 eq {}} {
                    # puts $comb
                    incr fitsmin
                }
            }
        }
        if {$fitsmin != 0} {
            set result2 $fitsmin
        }
    }
    yield $fits
    yield $result2
}
```
