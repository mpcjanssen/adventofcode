```racket
(require racket/file)
(require threading)
(define (delta char)
      (if (equal? char #\( ) 1 -1))

(define deltas (~>
     "input/1.txt"
     file->string
     string->list
     (map delta _)))

(define part1
    (apply + deltas))

(define-values (part2 floor)
   (for/fold ([idx 0] [floor 0]) ([d deltas]) 
        #:break (< floor 0)
        (values (+ 1 idx) (+ floor d))))

(list part1 part2)
```
